```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = 'school_grades.csv'
data = pd.read_csv(file_path);
df = data;
```

```python
print('Skewness ', df['math score'].skew());
print('Mean ', np.mean(df['math score']))
print('Median ', np.median(df['math score']))
print('Mode ', df['math score'].mode())
```

```python
df.corr(numeric_only=True)
```

```python
colors = ['red','blue']
df.plot.scatter(x='reading score',y='writing score')
```

```python
pivottable = df.pivot_table(index='class group', columns='gender', values=['reading score', 'math score'], aggfunc=['mean']).round(2)
pivottable.plot(kind='line')
```
